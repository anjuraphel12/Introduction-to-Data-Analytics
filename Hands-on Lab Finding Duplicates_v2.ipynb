## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
## Write your code here
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")
duplicate_rows = df[df.duplicated()]
print(duplicate_rows.head())
```

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
## Write your code here
# Identify duplicates based on selected columns
duplicate_subset = df[df.duplicated(subset=['MainBranch', 'Employment', 'RemoteWork'], keep=False)]
# Display duplicate rows
print(f"Number of duplicate rows based on selected columns: {duplicate_subset.shape[0]}")
print(duplicate_subset.head())  # Show first few duplicate rows


# Identify columns where values are identical across duplicate rows
identical_columns = (duplicate_subset.nunique() == 1)  # Columns with only 1 unique value in duplicates
# Display columns that are frequently identical
frequent_identical_cols = identical_columns[identical_columns].index.tolist()
print(f"Columns that frequently contain identical values in duplicates: {frequent_identical_cols}")


# Count occurrences of each duplicate subset
duplicate_summary = duplicate_subset.groupby(['MainBranch', 'Employment', 'RemoteWork']).size().reset_index(name='Count')
# Display the top duplicate patterns
print("Top duplicate patterns based on MainBranch, Employment, and RemoteWork:")
print(duplicate_summary.sort_values(by='Count', ascending=False).head(10))
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
## Write your code here
# Identify duplicates based on all columns
duplicate_rows = df[df.duplicated(keep=False)]
# Count duplicate rows by Country
country_counts = duplicate_rows['Country'].value_counts().head(10)  # Top 10 countries with duplicates

# Plot
# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(country_counts.index, country_counts.values, color='skyblue')
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Number of Duplicates')
plt.title('Top 10 Countries with the Most Duplicates')
plt.show()
```

```python
# Count duplicates by Employment type
employment_counts = duplicate_rows['Employment'].value_counts()

# Plot
plt.figure(figsize=(8, 8))
plt.pie(employment_counts, labels=employment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Duplicates by Employment Type')
plt.show()
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
## Write your code here
# Check for duplicate rows based on all columns
print(f"Total duplicate rows: {df.duplicated().sum()}")

# Check duplicates based on selected columns
selected_cols = ['MainBranch', 'Employment', 'RemoteWork']
print(f"Duplicates based on {selected_cols}: {df.duplicated(subset=selected_cols).sum()}")

# Drop duplicates based on selected columns
df_deduplicated = df.drop_duplicates(subset=selected_cols, keep='first')
# Display results
print(f"Rows after removing duplicates: {df_deduplicated.shape[0]}")
```
